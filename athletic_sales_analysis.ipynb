```python
# Import Libraries and Dependencies
import pandas as pd
import os as os
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.

#path
file_path="./Resources/"

#get a list of csv files in file_path
file_list=[file for file in os.listdir(file_path) if file.endswith('.csv')]

#loop through file_list to import files
dataframes_dict={}
for i in range(len(file_list)):
    dataframes_dict[file_list[i][:len(file_list[i])-4]]=pd.read_csv(file_path+file_list[i])
    #print(i)

print(dataframes_dict.keys())
print(type(dataframes_dict['athletic_sales_2020']))
print(type(dataframes_dict['athletic_sales_2021']))



```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.

#checks on the dataframes
#check the shape
print("2020 df shape: ",dataframes_dict['athletic_sales_2020'].shape)
print("2021 df shape: ",dataframes_dict['athletic_sales_2021'].shape)
#check if both dataframes have the same number of columns in the same order
print("columns the same? " ,dataframes_dict['athletic_sales_2020'].columns.equals(dataframes_dict['athletic_sales_2021'].columns))

#check if the dytpes of the columns are the same
print("dtypes the same? ",dataframes_dict['athletic_sales_2020'].dtypes.equals(dataframes_dict['athletic_sales_2021'].dtypes))

#combine the dataframes
all_sales=pd.merge(dataframes_dict['athletic_sales_2020'],dataframes_dict['athletic_sales_2021'], how='outer')

#check that the shape makes sense
print("merged df shape",all_sales.shape)

#look at data
all_sales
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".
num_sold_groupby=all_sales.groupby(["region","state","city"])[["units_sold"]].sum()

num_sold_groupby=num_sold_groupby.rename(columns={"units_sold":"Total_Products_Sold"})

# Show the top 5 results.
num_sold_groupby=num_sold_groupby.sort_values("Total_Products_Sold", ascending=False)

num_sold_groupby.head(5)
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.
sales_groupby=all_sales.groupby(by=["region","state","city"])[["total_sales"]].sum()

# Rename the "total_sales" column to "Total Sales"
sales_groupby=sales_groupby.rename(columns={"total_sales":"Total Sales"})

# Show the top 5 results.
sales_groupby=sales_groupby.sort_values("Total Sales", ascending=False)

sales_groupby.head()
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.
sales_pt=pd.pivot_table(data=all_sales,values="total_sales",index=["region","state","city"],aggfunc="sum")

# Optional: Rename the "total_sales" column to "Total Sales"
sales_pt=sales_pt.rename(columns={"total_sales":"Total Sales"})

# Show the top 5 results.
sales_pt=sales_pt.sort_values("Total Sales", ascending=False)

sales_pt.head()
```
